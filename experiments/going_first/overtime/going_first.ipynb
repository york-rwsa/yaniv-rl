```python
import numpy as np
import matplotlib.pyplot as plt
import json
import matplotlib.ticker as ticker
```

```python
with open("./going_first_over_time.json") as f:
    results = json.load(f)
```

```python
wins = [r['player_0_win_mean'] for r in results]
iteration = [r['model_number'] for r in results]

plt.ylabel('avg win-rate')
plt.xlabel('training iteration')
plt.title('avg win-rate over 20000 games vs training iteration')
plt.scatter(iteration, wins, s=3)
```
