```python
import numpy as np
import matplotlib.pyplot as plt
import json
import matplotlib.ticker as ticker
```

```python
with open("./intermediate_vs_models.json") as f:
    results = json.load(f)
```

```python
wins = [r['player_0_win_mean'] for r in results]
iteration = [r['model_number'] for r in results]

plt.ylabel('avg win-rate')
plt.xlabel('training iteration')
plt.title('avg win-rate over 100000 games vs training iteration')
plt.scatter(iteration, wins, s=3)
```
