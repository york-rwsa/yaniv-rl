```python

cats = ['5S', '4K', '4S', '3S2K', '3K2K', '3S', '3K', '2K2K', '2K', 'X']
x = np.arange(len(cats))
winrates = [res[c]['wins'][0] / 10000 for c in cats]
width = 0.3

fig, ax = plt.subplots()

ax.bar(x - width / 2, [r['player_0_win_mean'] for r in model_res], width, color="tab:blue", label="Best RL Model")
ax.bar(x + width / 2, winrates, width, color="tab:orange", label="Intermediate Rule Agent")

ax.axhline(linestyle=":", y=0.5, color="black")

ax.legend(ncol=2, loc='lower center')

ax.set_xticks(x)
ax.set_xticklabels(cats)
ax.set(title="RL Model and Intermediate Agent Self Play per starting hand class", xlabel="Hand Class", ylabel="Avg. Win Rate")

fig.tight_layout()
```
