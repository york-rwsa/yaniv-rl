```python
import numpy as np
import matplotlib.pyplot as plt
import json
import matplotlib.ticker as ticker
```

```python
with open("./handclassoutput_20000_every_100.json") as f:
    results = json.load(f)
```

```python
models = sorted(map(int, results.keys()))

classes = ['5S', '4K', '4S', '3S2K', '3K2K', '3S', '3K', '2K2K', '2K', 'X']

fig, ax = plt.subplots()

for i, c in enumerate(classes):
    winrates = [results[str(m)][i]["player_0_win_mean"] for m in models]
    plt.plot(models, winrates, label=c)

ax.set(title="Win Rate of hand class over training iteration", xlabel="Training Iteration", ylabel="Avg. Win Rate (over 20000 games)")

leg = ax.legend(loc='lower center', ncol=5)
```
